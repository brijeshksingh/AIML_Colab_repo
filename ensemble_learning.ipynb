```python
# prompt: getting this "''''''''" in above sql string output

# The previous cell modified the dataframe 'df' by applying sanitize_sql_string
# with .strip("'"). This meant the data in the dataframe itself was changed.
# The generate_insert_statement function expects the original data to apply
# the sanitization within the function call to correctly format it for SQL.

# To fix this, we need to reload or reset the dataframe to its state
# before the cell that modified it by stripping quotes. Since we don't
# have the original state easily accessible here, the simplest fix is to
# remove the dataframe modification part and ensure the sanitize_sql_string
# function is correctly defined and used *only* within the
# generate_insert_statement function.

# Let's redefine the sanitize_sql_string function and the
# generate_insert_statement function to ensure they work correctly together.
# We will assume 'df' still holds the data from the previous steps *before*
# the last cell that modified the dataframe content itself. If you ran the
# last cell, you might need to re-run the notebook up to the cell before it
# or reload the dataframe from the original source to get the correct state.

def sanitize_sql_string(value):
    """
    Sanitizes a value for use in a SQL string literal.
    Handles NULL/NaN, escaping single quotes, and cleaning newlines.
    """
    if pd.isna(value):
        return "NULL"
    # Convert non-string types to string
    value_str = str(value)
    # Escape single quotes by doubling them
    value_str = value_str.replace('\\ Tax\\', " Tax ")
    # Replace newlines with spaces or remove them
    value_str = value_str.replace("'", "\"")
    # Replace newlines with spaces or remove them
    value_str = value_str.replace('\n', ' ').replace('\r', '')
    # Wrap the sanitized string in single quotes
    return f"'{value_str}'"

def generate_insert_statement(df, project_id, dataset_id, table_name):
    """
    Generates a BigQuery INSERT INTO statement from a pandas DataFrame.

    Args:
      df: pandas DataFrame.
      project_id: The Google Cloud project ID.
      dataset_id: The BigQuery dataset ID.
      table_name: The BigQuery table name.

    Returns:
      A string representing the BigQuery INSERT INTO statement.
    """
    if df.empty:
        return ""

    columns = ", ".join([f"`{col}`" for col in df.columns])

    values_list = []

    for index, row in df.iterrows():
        # Sanitize each value for SQL insertion *within* this loop
        row_values = [sanitize_sql_string(row[col]) for col in df.columns]
        values_list.append(f"({', '.join(row_values)})")

    insert_statement = f"""INSERT INTO `{project_id}.{dataset_id}.{table_name}` ({columns}) VALUES
{", ".join(values_list)};""" # Use ",\n" for better readability of the generated SQL

    return insert_statement

# Example usage:
# Replace with your actual project ID, dataset ID, and table name
project_id = "your_project_id"
dataset_id = "abc_test"
table_name = "your_target_table_name"

# Assuming 'df' still holds the data correctly
sql_insert_script = generate_insert_statement(df, project_id, dataset_id, table_name)
print(sql_insert_script) # Use print() to see the generated string without extra quotes

```
